## **Praktek: Merubah dan menghapus elemen data pada array​**

```python
#import library numpy
import numpy as np
```

```python
#inisiasi array A
A = np.array([1, 3, 6, 3])
```

```python
#merubah elemen array A yang >5 menjadi 0
A[A > 5] = 0
```

```python
np.sqrt(A)
```

```python
#print hasil array yang telah dirubah
print(A)
import math
np.emath.sqrt(A)
```

```python
A
```

```python
#menghapus elemen A pada indeks ke-0 sampai batas indeks 3 dengan np.delete(nama_array, indeks)
print(A)
A1 = np.delete(A, np.where(A[0:3]))
print(A1)
```

```python
#print hasil array yang telah dihapus elemen pertamanya (elemen dengan indeks = 0)
print(A1)
```

## **Praktek: Melakukan pengurutan dan penyaringan data pada array​**

```python
#import library numpy
import numpy as np
```

```python
#inisiasi array A
A = np.array([1, 3, 6, 3])
```

```python
#urutkan array A
A2 = np.sort(A)
```

```python
#print hasil array yang telah diurutkan
print(A2)
```

```python
#filter elemen array berdasarkan indeks
print(A2)
x = [True, False, True, False]
A3 = A2[x]
```

```python
#print hasil array yang telah difilter
print(A3)
```

```python
#Filtering Array based on conditions
#buat filter yang hanya angka>3 yang disertakan
filter_arr = A > 3
A4 = A[filter_arr]
```

```python
#print hasil array yang telah difilter
print(A4)
A4.dtype
```
