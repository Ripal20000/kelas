## **Praktek: filter data pada List​**

```python
#inisiasi list L
L = [1, 3, 6, 3]
```

```python
#tambahkan elemen ke dalam list L, simpan dengan nama L1
L1 = L + [0, 8, 2]
L1
```

```python
#filter list L1
L2 = list(filter(lambda L1: L1 > 5, L1)) #def function
L2
```

## **Praktek: Menggunakan nested list untuk data dua dimensi​**

```python
#inisiasi data dua dimensi dengan nested list
produk = [['Balsem Putih 20 gram', 13],
          ['Batere AA', 184],
          ['Bawang Putih 200 gram', 402],
          ['Beras Pulen 5 kg', 27],
          ['Biskuit Waver Kaleng', 135],
          ['Gula Pasir Putih 1 kg', 547],
          ['Ketan Hitam 500 gram', 23],
          ['Kopi Instant', 384],
          ['Mi Goreng Instant', 642],
          ['Obat Batuk', 73],
          ]
```

```python
terlaris = list(filter(lambda c:c[1] > 200, produk))
```

```python
terlaris
```

```python
for x in terlaris:
  filter_produk = x[0]
  print(filter_produk)
```

```python
for x in range(len(terlaris)):
  print(terlaris[x][0])
```

```python
for cust in produk:
    print(f"Nama Produk = {cust[:1]}")
```

```python
for x in produk:
  nama_produk = x[0]
  print("Nama Produk:",nama_produk)
```

## **Praktek: Menggunakan nested list untuk data lebih dari dua dimensi​**

```python
produk = [['Balsem Putih 20 gram', 22300, 13],
          ['Batere AA', 12000, 184],
          ['Bawang Putih 200 gram', 15000, 402],
          ['Beras Pulen 5 kg', 63000, 27],
          ['Biskuit Waver Kaleng', 24000, 135],
          ['Gula Pasir Putih 1 kg', 12500, 547],
          ['Ketan Hitam 500 gram', 18000, 23],
          ['Kopi Instant', 5200, 384],
          ['Mi Goreng Instant', 2600, 642],
          ['Obat Batuk', 5000, 73],
          ]
```

```python
terlaris = filter(lambda c: (c[1]>5000 and c[2]>200), produk)
```

```python
print(list(terlaris))
```
